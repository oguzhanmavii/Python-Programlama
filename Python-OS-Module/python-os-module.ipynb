## os.stat
os modülünün stat() fonksiyonu dosyalar hakkında bilgi almamızı sağlar. Bu fonksiyonu kullanarak bir dosyanın boyutunu, oluşturulma tarihini, değiştirilme tarihini ve erişilme tarihini sorgulayabiliriz.

```python
import os
dosya=os.stat('/home/jupyter-tarkanoguzhan/python-os-module.ipynb')
print(dosya)
```

Burada özellikle işimize yarayacak olan nitelikler şunlardır:

st_atime
dosyaya en son erişilme tarihi

st_ctime
dosyanın oluşturulma tarihi (Windows’ta)

st_mtime
dosyanın son değiştirilme tarihi

st_size
dosyanın boyutu

Bu metinde açıkladıgımız bütün parametreler yukarıdakiçıktıda yer alıyor inceleyebilirsiniz

Not:⚠GNU/Linux’ta bir dosyanın ne zaman oluşturulduğunu öğrenmek mümkün değildir. Dolayısıyla dosya.st_ctime komutu yalnızca Windows’ta bir dosyanın oluşturulma tarihi verir. Bu komutu GNU/Linux’ta verdiğimizde elde edeceğimiz şey dosyanın son değiştirilme tarihidir.

## os.walk
os modülü içindeki listdir() adlı bir fonksiyondan söz etmiştik. Bu fonksiyon, bir dizinin içeriğini listeleme imkanı veriyordu bize. Mesela o anda içinde bulunduğumuz dizinde hangi dosya ve alt dizinlerin olduğunu öğrenmek için şöyle bir komut kullanabiliyorduk:os.listdir('.')
simdi ise bir adet test adinda klasor  olusturalim icerisine testing.py isminde bir python dosyası olusturalim icersine asagidaki kodlari yazalim <br>

```python
import os

uzantılar = ['txt', 'doc', 'xls',
             'jpeg', 'pdf', 'zip',
             'mp3', 'ogg', 'jpeg']

şablon1 = ['{}.{}'.format('dosya', i) for i in uzantılar[:4]]
şablon2 = ['resim{}.{}'.format(i, uzantılar[-1]) for i in range(1, 5)]
şablon3 = ['{}.{}'.format('dosya', i) for i in uzantılar[4:]]

dosyalar = [('anadizin',  şablon1),
            ('resimler', şablon2),
            ('başkadosyalar', şablon3)]

os.makedirs(os.sep.join([dosya[0] for dosya in dosyalar]))

for dizin, şablon in dosyalar:
    for s in şablon:
        open(os.sep.join([dizin, s]), 'w')
    os.chdir(dizin)
```

Terminal ekranina __python testing.py__ yazip enter bastigimizda <br>
<br>
__anadizin__ isminde bir klasör oluşturulacaktır

daha sonra test klasörümüze controller.py isminde dosya oluşturuyoruz
<br>
<br>
şimdi ise controller.py dosyasini oluşturalim içerisine asagidaki kodu yazalim

```python
import os
for i in os.walk('anadizin'):
    print(i)
```

Terminal ekranina __python controller.py__ yazip enter bastigimizda <br>
Terminal Ekranımızda  şu çıktıyı alacağız:

('anadizin', ['resimler'], ['dosya.doc', 'dosya.jpeg',
              'dosya.txt', 'dosya.xls'])
('anadizin\\resimler', ['başkadosyalar'], ['resim1.jpeg',
              'resim2.jpeg', 'resim3.jpeg', 'resim4.jpeg'])
('anadizin\\resimler\\başkadosyalar', [], ['dosya.jpeg',
              'dosya.mp3', 'dosya.ogg', 'dosya.pdf', 'dosya.zip'])

Gördüğünüz gibi, burada üç öğeli bir demet var. Çıktının diğer kısımlarını da incelerseniz aynı yapıyı göreceksiniz. Dolayısıyla __os.walk()__ komutu bize şu üç öğeden oluşan bir demet verir: (kökdizin, altdizinler, dosyalar)

## os.environ
os modülünün environ adlı niteliği, kullandığımız işletim sistemindeki çevre değişkenleri hakkında bilgi edinmemizi sağlar.
Bu nitelik alelade bir sözlüktür. Dolayısıyla bu sözlüğün içinde neler olduğunu aşağıdaki kodlarda görebilirsiniz

```python
import os 
for k,v in os.environ.items():
    print(k.ljust(10),v)
```

Bunlardan bazılarına değinelim <br>
<li>os.path.abspath() fonksiyonu, bir dosyanın tam yolunun ne olduğunu söyler</li>
<li>os.path.dirname() fonksiyonu, bir dosya yolunun dizin kısmını verir </li>
<li>os.path.exists() fonksiyonu bir dosya veya dizinin varolup olmadığını kontrol eder</li>
<li>os.path.expanduser() fonksiyonu bilgisayardaki kullanıcıya ait dizinin adresini verir</li>
<li>os.path.isdir() fonksiyonu, kendisine parametre olarak verilen öğenin bir dizin olup olmadığını sorgular</li> 
<li>os.path.isfile() fonksiyonu, kendisine parametre olarak verilen öğenin bir dosya olup olmadığını sorgular</li>
<li>os.path.join()fonksiyonu, kendisine verilen parametrelerden, ilgili işletim sistemine uygun yol adresleri oluşturur</li>
<li>os.path.split() fonksiyonu, bir yol adresinin son kısmını baş kısmından ayırır</li>
<li>os.path.splitext() fonksiyonu dosya adı ile uzantısını birbirinden ayırmak için kullanılır</li>
